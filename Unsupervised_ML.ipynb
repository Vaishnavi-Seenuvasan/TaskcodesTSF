```
# Importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
#importing the dataset
dataset = pd.read_csv('/content/Iris (1).csv')
x = dataset.iloc[:, [1, 2, 3, 4]].values
```

```
dataset.head(10) #view the first 10 rows
```

```
#executing elbow method to find optimum clusters
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11): #Choosing this range as it is easy to work with but no matter what the range is elbow remains constant
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```
plt.plot(range(1, 11), wcss)
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('cluster sum of squares') #gives within clusters the sum of squares
plt.show()#plots the result to get the elbow value
```

```
#Application of kmeans clustering/ Creating the kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```
#Visualise the clusters to gain an overall idea 
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
#Centroids are also plotted
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
