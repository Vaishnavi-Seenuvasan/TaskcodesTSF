```
#importing the libraries needed
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
#importing the data
data=pd.read_csv("/content/SampleSuperstore.csv")
```

```
#cleaning data
data.isnull().sum() #finding null cells
```

```
store=data.drop(['Country', 'Postal Code','Ship Mode'], axis=1) #dropping a few columns to clean data
```

```
store.head() #data after dropping unnecessary columns
```

```
#relationship analysis
correlation=store.corr()
```

```
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True) 
```

```
sns.pairplot(store) #pairwise relation but not intuitive and clear to comprehend
```

```
sns.barplot(x='Region', y='Sales', hue='Segment', data=store) #analysing in a lucid form
#it can be observed that home office is largest segment overall followed by corporate and consumer
#concentrate more on the major sales segment to increase profits
```

```
sns.barplot(x='Discount', y='Sales', hue='Category', data=store) #Discount in technology gives more sales work with that and widen the strategy to other categories as well
```

```
sns.distplot(store['Profit']) #overall distribution of profits
```

```
sns.distplot(store['Sales']) #overall distribution of sales
```

```
sns.catplot(x='Quantity', hue='Category', kind='count', data=store) #number items sold in various categories
#Office supplies have sold more comparatively. Hence stock up more in order to augment sales and hence profits 
```

```
#find which sub-category items are sold more
plt.figure(figsize=(40,40))
store.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Sales and profit in each sub-category') 
#chairs and phone have brought in more sales and profit than the rest of the sub-category items
plt.show()
```
